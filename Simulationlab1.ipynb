```python
#importing the  necessary module
import numpy as np  # array manipulation
import matplotlib.pyplot as plt  #  graph plotting
from scipy.stats import poisson  # scientifi  statistical distribution
k=np.arange(0,17)  # random number (return an array)
pmf= poisson.pmf(k,mu=7)
pmf=np.round(pmf,5)
plt.plot(k, pmf,color='m',marker='o')
plt.xlabel('k')
plt.ylabel("Probability")
plt.show()
```

```python
cdf=poisson.cdf(k,mu=7)
cdf=np.round(cdf,3)
plt.plot(k,cdf,color='m',marker='o')
plt.xlabel('ksds')
plt.ylabel('Cumulative  Jiban Probability')
plt.show()
```
